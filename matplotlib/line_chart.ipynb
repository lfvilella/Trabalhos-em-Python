```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
precipitacao = [temps for temps in np.random.randint(0,50,size=12)]
precipitacao
```

```python
plt.plot(precipitacao)
plt.xticks(ticks=np.arange(len(precipitacao)), labels=months); #this ';' is to simplify, just remove and see
```

```python
low_precip = np.argmin(precipitacao)
high_precip = np.argmax(precipitacao)
```

```python
plt.plot(precipitacao)
plt.xticks(ticks=np.array([low_precip, high_precip]), labels=months[[low_precip, high_precip]])
```
