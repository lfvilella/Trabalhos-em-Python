```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(np.random.randint(0,10,10), label='One')
plt.plot(np.random.randint(0,10,10), label='Two')
plt.plot(np.random.randint(0,10,10), label='Tree')
plt.plot(np.random.randint(0,10,10), label='Four')
plt.plot(np.random.randint(0,10,10), label='Five')
plt.legend(loc=8, ncol=2, facecolor='#FFCCCC', fontsize='large', shadow=True, title='Title');
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Agu', 'Sep', 'Oct']
wedgets, _ = plt.pie(np.random.randint(0,10,10));
plt.legend(wedgets, months, bbox_to_anchor=(1.0, 1.0))
```
