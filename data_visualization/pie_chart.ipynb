```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

first_values = np.array([num for num in np.random.randint(0, 20, 12)])
second_values = np.array([num for num in np.random.randint(0, 20, 12)])
third_values = np.array([num for num in np.random.randint(0, 20, 12)])
```

```python
plt.pie(first_values, labels=months, autopct='%.2f%%', rotatelabels=True, startangle=90);
```

```python
october_data = [value[9] for value in [first_values, second_values, third_values]]
```

```python
some_labels = ['2020', '2019', '2018']
```

```python
_, years, pcts = plt.pie(october_data, autopct='%.2f%%', labels=some_labels) # without ';'
years[0].set_fontweight('bold')
pcts[1].set_fontstyle('italic')
```

```python
explosions = np.zeros(12)

min_position = np.argmin(first_values)
explosions[min_position] = 0.6

max_position = np.argmax(first_values)
explosions[max_position] = 0.3

plt.pie(first_values, labels=months, explode=explosions);
```
