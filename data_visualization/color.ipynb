```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-np.pi * 2, np.pi * 2, 721)
y = np.sin(x)
for offset, color in zip(np.arange(8), list('rgbcmykw')):
    plt.plot(x, y + offset, c=color)
```
