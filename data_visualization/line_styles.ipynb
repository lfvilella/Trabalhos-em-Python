```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
styles = ['solid', 'dashed', 'dashdot', 'dotted', 'None']
```

```python
for idx, style in enumerate(styles):
    plt.plot(np.arange(10) + idx, linestyle=style)
```

```python
abbreviations = ['-', '--', '-.', ':', '']
for idx, abbreviations in enumerate(abbreviations):
    plt.plot(np.arange(10) + idx, ls=abbreviations)
```

```python
plt.plot(np.arange(10), ls=(0, (1,10,5,1)), linewidth=10)
plt.plot(np.arange(10)+1, ls=(0, (1,10,5,1)))
```

```python
plt.plot(np.arange(10), ls='--', linewidth=10)
plt.plot(np.arange(10) + 1, ls='--', linewidth=10, dash_capstyle='round')
```

```python
plt.plot(np.random.randint(0,10,10), lw=10, solid_joinstyle='miter')
```
