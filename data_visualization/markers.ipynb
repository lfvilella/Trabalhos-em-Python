```python
points = np.array(list(zip(np.random.uniform(size=100), np.random.uniform(size=100))))
```

```python
quadrants = {
    1: [],
    2: [],
    3: [],
    4: [],
}
```

```python
for point in points:
    if point[0] >= 0.5 and point[1] >= 0.5:
        quadrants[1].append(point)
        
    if point[0] < 0.5 and point[1] >= 0.5:
        quadrants[2].append(point)
        
    if point[0] < 0.5 and point[1] < 0.5:
        quadrants[3].append(point)
        
    if point[0] >= 0.5 and point[1] < 0.5:
        quadrants[4].append(point)
```

```python
for k in quadrants.keys():
    quadrants[k] =  np.array(quadrants[k])
```

```python
plt.scatter(quadrants[1][:,0], quadrants[1][:,1], marker=(3,0,75), s=np.ones(len(quadrants[1]))*100)
plt.scatter(quadrants[2][:,0], quadrants[2][:,1], marker=(3,0,165), s=np.ones(len(quadrants[2]))*100)
plt.scatter(quadrants[3][:,0], quadrants[3][:,1], marker=(3,0,255), s=np.ones(len(quadrants[3]))*100)
plt.scatter(quadrants[4][:,0], quadrants[4][:,1], marker=(3,0,345), s=np.ones(len(quadrants[4]))*100)
```
