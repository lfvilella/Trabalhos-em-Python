```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

first_values = np.array([num for num in np.random.randint(0, 20, 12)])
second_values = np.array([num for num in np.random.randint(0, 20, 12)])
third_values = np.array([num for num in np.random.randint(0, 20, 12)])
```

```python
plt.bar(np.arange(len(months)), [value for value in first_values]); 
plt.xticks(ticks=np.arange(len(months)), labels=months);
plt.title('The Title Here', fontdict={
    'size': 'x-large',
    'weight': 'bold',
});
plt.ylabel('Some metry');
plt.xlabel('Month');
plt.text(8, 15.3, '15');
```

```python
plt.barh(np.arange(len(months)), [value for value in first_values]); 
plt.yticks(ticks=np.arange(len(months)), labels=months);
```

```python
plt.bar(np.arange(len(months)), first_values); 
plt.bar(np.arange(len(months)), second_values, bottom=first_values); 
plt.bar(np.arange(len(months)), third_values, bottom=first_values + second_values);
plt.xticks(ticks=np.arange(len(months)), labels=months);
```

```python
group_width = 0.9
```

```python
plt.bar(np.arange(len(months)) - group_width/3, first_values, group_width/3)
plt.bar(np.arange(len(months)), second_values, group_width/3)
plt.bar(np.arange(len(months)) + group_width/3, third_values, group_width/3)
plt.xticks(ticks=np.arange(len(months)), labels=months);
```
